This is a preliminary exploration of the sa_reports_0615.csv data.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
import pandas as pd
df = pd.read_csv('sa_reports_0615.csv')
print('Number of variables: ' + str(len(df.columns)))
```

I mainly analyze the quantitative variables (quant_var) and categorical variables (cat_var) with the greatest number of non-zero data entries.

```python
quant_var = ["Wage Expenditure (Unskilled + Skilled) (Rs.): ", 
    "Material Expenditure (Rs.): ", 
    "Total Expenditure (Rs.): ",
    "Wage Records Given by Implementing Agency (Rs.): ",
    "Material Records Given by Implementing Agency (Rs.): ",
    "Total Records Given by Implementing Agency (Rs.): ",
    "Total House Holds Worked : ",
    "Total House Holds Verified: ",
    "Number of people participated in gram sabha: ",
    "Total Expenses: "]
cat_var = ["Are Job cards with people : ",
           "Are Job cards updated : ",
           "Have Job cards been renewed : ",
           "Is there a process to register demand and give receipts : ",
           "Is there a demand for work that is not met : ",
           "Do labourers have problems getting wages from the payment agency : ",
           "Are muster rolls maintained in the worksite : ",
           "Are wage slips given to people : ",
           "Is Rozgar Diwas was held every month : ",
           "Are the seven registers being maintained at the gram panchayat level : ",
           "Have citizen information boards been erected for community works : ",
           "Have citizen information boards been erected for individual and household works : ",
           "Have 'Wall writings' been done in the panchayat : ",
           "Is there a grievance redressal process : ",
           "Did the gram sabha approve a prioritized list of projects to be taken up under MGNREGA : ",
           "If yes to previous question, were the works taken up as per the priority : ",
           "Is drinking water provided at the worksite : ",
           "Is chaya/shade provided at the worksite : ",
           "Is first aid kit available at the worksite : ",
           "Is women worker been appointed in places with more than 5 children :",
           "Have all mates been selected through gram sabha/state norms : ",
           "Have mates been trained : ",
           "Is there adequate manpower to implement MGNREGA at panchayat level : ",
           "Is there a person in charge of MGNREGA at the panchayat level (not holding additional charge) : ",
           "Are the personnel in charge of MGNREGA at the panchayat level been trained : ",
           "Does the gram panchayat have adequate technical support personnel : "]
```

Summary statistics of the quantitative variables:

```python
df[quant_var].describe()
```

Shorten the variable names

```python
rename_dict = {"Wage Expenditure (Unskilled + Skilled) (Rs.): " : "Wage Expenditure", 
    "Material Expenditure (Rs.): " : "Material Expenditure", 
    "Wage Records Given by Implementing Agency (Rs.): " : "Wage Records",
    "Material Records Given by Implementing Agency (Rs.): " : "Material Records",
    "Total Records Given by Implementing Agency (Rs.): " : "Total Records",
    "Total House Holds Worked : " : "Total House Holds Worked",
    "Total House Holds Verified: " : "Total House Holds Verified",
    "Number of people participated in gram sabha: " : "Number of participants",
    "Total Expenses: " : "Total Expenses"}
df_quant = df[quant_var].rename(columns=rename_dict)[rename_dict.values()]
df = df.rename(columns=rename_dict)
df.head()
```

Correlation heatmap of the quantitative variables

```python
fig, ax = plt.subplots()
sns.heatmap(
    df_quant.corr(),
    ax=ax,
    cmap="viridis",
)
ax.set_title("Quantitative Variables Correlation Heatmap")
ax.set_xticklabels(
    ax.get_xticklabels(), rotation=20, ha="right", rotation_mode="anchor"
)
#fig.tight_layout()
#fig.savefig(os.path.join(FIGURE_DIR, "metric_score_corr.png"), dpi=300)
```

Categorical variable percentages

```python
for cat in cat_var:
  print(cat)
  print(df[cat].value_counts(normalize=True))
  print('\n')
```

Summary statistics of wage expenditure grouped by categorical variables

```python
for cat in cat_var:
    print(df[['Wage Expenditure', cat]].groupby(cat).describe())
    print('\n')
```

```python
for cat in cat_var:
    print(df[['Material Expenditure', cat]].groupby(cat).describe())
    print('\n')
```



Summary statistics of total expenses grouped by the categorical variables



```python
for cat in cat_var:
    print(df[['Total Expenses', cat]].groupby(cat).describe())
    print('\n')
```

Histogram of total expenses grouped by categorical variables

```python
for cat in cat_var:
    fig, ax = plt.subplots()
    sns.histplot(df, ax=ax, x='Total Expenses', hue=cat, kde=True)
    ax.set_xlim(0,1e5)
    plt.show()
```

Repeat the plot, but filter out zeros:

```python
filtered_df = df.loc[df['Total Expenses'] != 0]
for cat in cat_var:
    fig, ax = plt.subplots()
    sns.histplot(filtered_df, ax=ax, x='Total Expenses', hue=cat, kde=True)
    ax.set_xlim(0,1e5)
    plt.show()
```
